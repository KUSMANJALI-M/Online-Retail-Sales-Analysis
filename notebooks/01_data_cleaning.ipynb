```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv(r"D:\Ecommerce-Sales-Intelligence\data\online_retail_II.csv", encoding="ISO-8859-1")


```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

df.columns
```

```python
df = df.dropna(subset=['customer_id'])

df = df.dropna(subset=['description'])
```

```python
df['is_cancellation'] = df['invoice'].astype(str).str.startswith('C')

df = df[df['is_cancellation']==False]

df = df[df['quantity']>0]

df = df[df['price']>0]
```

```python
df['invoicedate'] = pd.to_datetime(df['invoicedate'])

df['customer_id'] = df['customer_id'].astype(int).astype(str)
```

```python
df['total_revenue'] = df['quantity'] * df['price']
```

```python
df['invoice_year'] = df['invoicedate'].dt.year
df['invoice_month'] = df['invoicedate'].dt.month
df['invoice_day'] = df['invoicedate'].dt.day
df['invoice_hour'] = df['invoicedate'].dt.hour

df['day_of_week'] = df['invoicedate'].dt.day_name()

df['year_month'] = df['invoicedate'].dt.to_period('M').astype(str)
```

```python
df.to_csv(r"D:\Ecommerce-Sales-Intelligence\data\cleaned_retail_data.csv",index=False)
print("Data Saved")
```
